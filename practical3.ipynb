```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.random.randint(1,100,size=(50,3)),columns=list('123'))
df.head(10)
```

```python
df2=pd.DataFrame(np.random.randint(700,800,size=(2,3)),columns=list('123'),index=[51,52])
df2
```

```python
df=pd.concat([df,df2],axis=0)
df
```

```python
for c in df.sample(int(df.shape[0]*df.shape[1]*0.10)).index:
    df.loc[c,str(np.random.randint(1,4))]=np.nan
```

```python
df.dropna(axis=1,thresh=df.shape[0]-5).head()
```

```python
#to drop the label of the max sum of the columns
label_max=df.sum(axis=1).idxmax()
print(label_max)
```

```python
df.drop(label_max,axis=0)
```
