```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('newdata.csv')
df
```

```python
df['Gender']=df['Gender'].str.strip()
df['Pass_Division']=df['Pass_Division'].str.strip()
df['Birth_Month']=df['Birth_Month'].str.strip()
df['Birth_Month']= df['Birth_Month'].astype('category')
```

```python
df['Gender']
df['Pass_Division']
```

```python
gender = pd.get_dummies(df['Gender'])
gender.head()
```

```python
div = pd.get_dummies(df['Pass_Division'])
div.head()
```

```python
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October",
          "November", "December"]
df['Birth_Month'] = pd.Categorical(df['Birth_Month'], categories=months, ordered=True)
df=df.sort_values('Birth_Month')
df.head()
```

```python
one_hot_encoding=pd.get_dummies(df,columns=['Gender','Pass_Division'])
```
